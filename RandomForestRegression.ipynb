```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
print(X_train)
print(Y_train)
print(X_test)
print(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(criterion='squared_error',max_features=None,n_estimators=100)
regressor=regressor.fit(X_train,Y_train)
```

```python
Y_predict=regressor.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,Y_predict)
r2
```

```python
import pickle
pickle.dump(regressor,open('finalized_RandomForest_model.sav','wb'))
```

```python
loaded_model=pickle.load(open('finalized_RandomForest_model.sav','rb'))
preinput=sc.transform([[40,28.5,2,1,0]])
result=loaded_model.predict(preinput)
print(preinput)
print(result)
```
